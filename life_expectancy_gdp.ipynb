```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("all_data.csv")
print(df.head())
```

```python
sns.barplot(data=df, x="Country", y="Life expectancy at birth (years)", ci="sd")
plt.show()
```

```python
sns.boxplot(data=df, x="Country", y="Life expectancy at birth (years)")
plt.show()
```

```python
df["GDP in Billions"] =df.GDP/1000000000
print(df.head())
```

```python
ax=plt.subplot()
sns.barplot(data=df, x="Country", y="GDP in Billions", ci="sd")
ax. set_xticklabels(["Chile", "China", "Germany", "Mexico", "USA", "Zimbabwe"], rotation=30)
plt.show()
```

```python
df_chile = df[df.Country == "Chile"]
df_china = df[df.Country == "China"]
df_germany = df[df.Country == "Germany"]
df_mexico = df[df.Country == "Mexico"]
df_usa = df[df.Country == "United States of America"]
df_zimbabwe = df[df.Country == "Zimbabwe"]
print(df_zimbabwe)

```

```python
year = list(df_chile["Year"])
chile_life_exp = list(df_chile["Life expectancy at birth (years)"])
china_life_exp = list(df_china["Life expectancy at birth (years)"])
germany_life_exp = list(df_germany["Life expectancy at birth (years)"])
mexico_life_exp = list(df_mexico["Life expectancy at birth (years)"])
usa_life_exp = list(df_usa["Life expectancy at birth (years)"])
zimbabwe_life_exp = list(df_zimbabwe["Life expectancy at birth (years)"])
```

```python


plt.plot(year, chile_life_exp, label = "Chile")
plt.plot(year, china_life_exp, label = "China")
plt.plot(year, germany_life_exp, label = "Germnay")
plt.plot(year, mexico_life_exp, label = "Mexico")
plt.plot(year, usa_life_exp, label = "USA")
plt.legend()
plt.xlabel("Year")
plt.ylabel("Life Expectancy in Years")
sns.set_context("notebook")
plt.savefig("Life Expectancy except Zimbabwe.png")
plt.show()
```

```python
sns.regplot(data=df_chile, x = "Life expectancy at birth (years)", y= "GDP in Billions", label = "Chile", color = "lightgreen")
plt.legend(loc=2)
plt.savefig("Chile GBP vs Life Exp.png")
plt.show()



```

```python
plt.plot(year, zimbabwe_life_exp, label = "Zimbabwe")
plt.legend()
plt.xlabel("Year")
plt.ylabel("Life Expectancy in Years")
sns.set_context("notebook")
plt.savefig("Life Expectancy Zimbabwe.png")
plt.show()
```

```python
sns.regplot(data=df_china, x = "Life expectancy at birth (years)", y= "GDP in Billions", label = "China", color = "blue")
plt.legend(loc=2)
plt.savefig("China GBP vs Life Exp.png")
plt.show()
```

```python
sns.regplot(data=df_germany, x = "Life expectancy at birth (years)", y= "GDP in Billions", label = "Germany", color = "orange")
plt.legend(loc=2)
plt.savefig("Germany GBP vs Life Exp.png")
plt.show()
```

```python
sns.regplot(data=df_mexico, x = "Life expectancy at birth (years)", y= "GDP in Billions", label = "Mexico", color = "purple")
plt.legend(loc=2)
plt.savefig("Mexico GBP vs Life Exp.png")
plt.show()
```

```python
sns.regplot(data=df_usa, x = "Life expectancy at birth (years)", y= "GDP in Billions", label = "USA", color = "darkblue")
plt.legend(loc=2)
plt.savefig("USA GBP vs Life Exp.png")
plt.show()
```

```python
sns.regplot(data=df_zimbabwe, x = "Life expectancy at birth (years)", y= "GDP in Billions", label = "Zimbabwe", color = "darkgreen")
plt.legend(loc=2)
plt.savefig("Zimbabwe GBP vs Life Exp.png")
plt.show()
```
